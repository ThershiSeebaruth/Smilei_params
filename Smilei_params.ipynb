Case n = 1e17 cm-3

In the simulation with 
n0 = 5e17
v0 =  1.8e7 #velocity in cm/s
the reflected beam reach 0.30 mm in 1 ns, meaning that n 10 ns it will reach (if no shock is created) 3 mm. 1.5 cm for the simulation box is way larger.


```python
#---------------------------------------------
# SIMULATION PARAMETERS FOR THE PIC-CODE SMILEI
# ---------------------------------------------

import math as m
import numpy as np
#
q  = 4.80326e-10       #statCoulomb
c  = 3.*1e10           #cm/s
me = 9.10938356*1e-28  #g
mi = 1836  #me/me = 1, mi/me = 1836



## INPUT ##
n0 = 1e17
# Eon plasma frequency in 1/s
v0 = 0.9e7 #velocity in cm/s
v0_c = v0/c

## ------- PISTON PLASMA -------

TeP_eV = 50 # temperature of electrons in eV
TiP_eV = 50 # temperature of ions in eV 
TeP_mec2 = TeP_eV/(511*1e3) # temperature nondimensionalization
TiP_mec2 = TiP_eV/(511*1e3) # temperature nondimensionalization

ZP = 6
neP = n0 #This actually has a profile, varies from 1e17 to 1e20 cm-3 
niP = neP/ZP

neP_n0 = neP/n0 #density nondimensionalization
niP_n0 = niP/n0 #density nondimensionalization

miP = mi*14 #Fully ionized Carbon




#--------Parameters for Input file----------------------------------------------------------------------
wpe = m.sqrt(4.*m.pi*n0*q**2/me)            # Eon plasma frequency in 1/s
c_wpe = c/wpe                               # Electron skin depth is reference length for Smilei), in cm.
inv_ld2 = (1/7.43e2)**2*(n0/TeP_eV)*(1+ZP)  # 1/lambda_D^2
deb_len = 1/np.sqrt(inv_ld2)                # Debye length

#------------- Case 1 -----------------------------------------------------------------------------------


Lx_rough = 1
cost =4
dx = cost*deb_len/c_wpe
Nx_rough = Lx_rough/c_wpe/dx


#--------------------------------------------------------------------------------------------------------

cell_per_patch = 8  #min 6
num_patches = 2**12 #power of 2 (prioritize large number of patches?)

#--------------------------------------------------------------------------------------------------------

Nx= num_patches*cell_per_patch 
Lx = dx*Nx # ~ 1.7 cm 
#Now, one should compare Lx with Lx_rough
print('Simulation box length (Nx*dx) :' +str(Lx*c_wpe))
print('')
print('Simulation box length (desired) :' + str(Lx_rough))
print('')
Nppc = int(cost/2*250) #Comes from considering the same amount of particles per length
#---------------------------------------------------------------------------------------------------------

# Normalization time in ns
dt = 0.9*dx  
Tsim_ns = 10 #Simulation time in ns
t_sim = int(Tsim_ns*wpe/1e9) #nondimensionalization 
Nt = t_sim/dt
push_time = 1000 #ns hopefully, a conservative value.

#---------------------------------------------------------------------------------------------------------


Expected_time_h = Nppc*Nt*Nx*push_time/1e9/3600
```

```python
print('Simulation box length (Nx*dx) :' +str(Lx*c_wpe))
print('')
print('Simulation box length (desired) :' + str(Lx_rough))
print('')
print('Nppc: ' +str(Nppc))
print('')
print('Nt: ' +str(Nt))
print('')
print('Nx: ' +str(Nx))
print('')
print('Note that this time should be divided by the # processors and the latter needs to be coherent with push time')
print('Expected time: ' + str(Expected_time_h) + ' h')
print('')
print('')
```
